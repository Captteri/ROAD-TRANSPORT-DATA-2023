```python
# Correlation matrix
# Select only numeric columns for correlation calculation
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(20,10))
sns.heatmap(correlation_matrix.round(2), annot=True)
plt.title('Correlation Matrix of Road Accidents Data')
plt.show()
```

```python
# Distribution of fatalities, serious injuries, and minor injuries
plt.figure(figsize=(12, 6))
sns.histplot(df['FATAL'], bins=20, kde=True, color='red', label='Fatalities')
sns.histplot(df['SERIOUS'], bins=20, kde=True, color='blue', label='Serious Injuries')
sns.histplot(df['MINOR'], bins=20, kde=True, color='green', label='Minor Injuries')
plt.title('Distribution of Fatalities, Serious Injuries, and Minor Injuries')
plt.xlabel('Number of Cases')
plt.legend()
plt.show()
```

```python
# Boxplot of fatalities by state
plt.figure(figsize=(16, 8))
sns.boxplot(x='STATE', y='FATAL', data=df)
plt.title('Boxplot of Fatalities by State')
plt.xlabel('State')
plt.ylabel('Number of Fatalities')
plt.xticks(rotation=45)
plt.show()
```

```python
import statsmodels.api as sm

# Define independent variables (factors)
independent_vars = ['SPV', 'DAD', 'BRD']  # Speed violations, alcohol involvement, road infrastructure

# Add constant term
df['constant'] = 1

# Define dependent variable (accident severity)
dependent_var = 'FATAL'  # You can replace 'FATAL' with the appropriate severity measure

# Fit Negative Binomial regression model
model = sm.GLM(df[dependent_var], df[independent_vars + ['constant']], 
               family=sm.families.NegativeBinomial()).fit()

# Print model summary
print(model.summary())
```

Speed Violations (SPV):
Coefficient (coef): 0.0020
Interpretation: For each unit increase in speed violations, the expected number of fatalities in road accidents increases by approximately 0.0020, holding other variables constant.
Statistical Significance: The coefficient is statistically significant (P-value < 0.05), indicating that speed violations have a significant impact on the number of fatalities.
Alcohol Involvement (DAD):
Coefficient (coef): -0.0056
Interpretation: For each unit increase in alcohol involvement, the expected number of fatalities in road accidents decreases by approximately 0.0056, holding other variables constant.
Statistical Significance: The coefficient is not statistically significant (P-value > 0.05), indicating that alcohol involvement may not have a significant impact on the number of fatalities in road accidents in this model.
Road Infrastructure (BRD):
Coefficient (coef): 0.0779
Interpretation: For each unit increase in road infrastructure issues (e.g., bad road conditions), the expected number of fatalities in road accidents increases by approximately 0.0779, holding other variables constant.
Statistical Significance: The coefficient is not statistically significant (P-value > 0.05), indicating that road infrastructure issues may not have a significant impact on the number of fatalities in road accidents in this model.
Constant Term:
Coefficient (constant): 3.5019
Interpretation: The expected number of fatalities in road accidents when all independent variables are zero is approximately 3.5019.
Statistical Significance: The constant term is statistically significant (P-value < 0.05), indicating that it contributes significantly to the model's prediction of fatalities.

Overall, the model suggests that speed violations have a significant positive impact on fatalities, while the influence of driving under alcohol/drug and bad road conditions is not statistically significant in predicting fatalities in road accidents.
